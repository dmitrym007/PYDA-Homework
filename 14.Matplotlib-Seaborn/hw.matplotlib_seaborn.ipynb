```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players_df = pd.read_csv('Players.csv')
players_df
```

```python
ax = players_df.height.hist(bins=25)
ax.set_title('Height distribution')
ax.set_xlabel('height')
ax.set_ylabel('count')
```

В гистограмме по оси абсцисс высота игрока, по оси ординат количество игроков.
По графику можно посмотреть распределение роста среди игроков.
Гистограмма в буквальном смысле это показывает (ヅ)☝

```python
sns.jointplot(x='height', y='weight', data=players_df, kind='reg')
plt.plot([160,240], [50, 130], color='g')
plt.text(175, 60, "Ideal weight by Broca's formula")
```

На графике joint_plot удобно оценить взаимосвязь высоты игрока(ось абсцисс) и веса(ось ординат) и ,например, сравнить линию регрессии с линией идеального веса.

```python
movies_df = pd.read_csv('tmdb_5000_movies.csv')
movies_df.head()
```

```python
movies_df['runtime'].plot(kind='box')
```

Для визуализации распределения длительности фильмов подходит график box plot. Из графика следует, что средняя длительность фильмов равна около 100 минут, а основная часть находится в пределах от 60 до 150 минут.

```python
revenue_df = movies_df.sort_values('revenue', ascending=False).iloc[:, [-3, -8]].head(15)
plt.barh(revenue_df['title'], revenue_df['revenue'])
plt.xlabel('Revenue')
plt.ylabel('Movie title')
plt.title('Top 15 movies with largest revenue')
plt.grid()
```

Для отображения распределения 15 наиболее доходных фильмов я воспользовался столбчатой диаграммой, т.к. она нагляднее показывает разницу. По оси абсцисс - доход, по оси ординат - названия фильмов.

```python
production_companies_dict = {}
for i in range(len(movies_df)):
    production_companies_dict[movies_df.original_title[i]] = pd.DataFrame(eval(movies_df.production_companies[i]))
```

```python
production_companies_df = pd.concat(production_companies_dict).reset_index(level=1, drop=True)
production_companies_df
```

```python
top_production_co = production_companies_df.name.value_counts().sort_values(ascending=False).reset_index().head(30)
```

```python
top_production_co.style.bar()
```

В стилизованной таблице числовые значения для наглядности дополняются столбчатой диаграммой.
