Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
```

```python
ratings_df = pd.read_csv('ml-latest-small/ratings.csv')
ratings_df['sum of rating 5'] = 1
ratings_df.head(3)
```

```python
movies_df = pd.read_csv('ml-latest-small/movies.csv')
movies_df.head(3)
```

```python
count_df = ratings_df[ratings_df.rating == 5].groupby('movieId').count().sort_values('rating', ascending=False)[['sum of rating 5']]
count_df.head(3)
```

```python
joined_df = count_df.merge(movies_df, on='movieId', how='right')[['title', 'sum of rating 5']]
joined_df.head()
```

Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
power_df = pd.read_csv('power.csv')
power_df.head(3)
```

```python
country_filter = '(country=="Latvia" or country=="Lithuania" or country=="Estonia")'
```

```python
year_filter = ' and year in ' + str(list(range(2005,2011)))
```

```python
category_filter = ' and category in [4, 12, 21]'
```

```python
filter_string = country_filter + year_filter + category_filter
filter_string
```

```python
power_df[power_df['quantity']>=0].query(filter_string).sum()
```
